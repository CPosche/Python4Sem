imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Making the blobs

```python
# create a random dataset with 2 features and 3 clusters
# X, y = make_blobs(n_samples=300, centers=3, random_state=42)
# can also be created with set features, although default is 2
# X, y = make_blobs(n_samples=1000, n_features=2, centers=3, random_state=42)
X, y = make_blobs(n_samples=300, n_features=2, centers=5, random_state=42)
# X, y = make_blobs(n_samples=300, n_features=2, centers=5, random_state=100)
```

standard scattered dataset

```python
# plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

clustering with DBSCAN

DBSCAN uses 2 parameters eps(the distance between points) and min_samples(the minimum number of points required to form a cluster)

```python
from sklearn.cluster import DBSCAN

# cluster the dataset using DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
y_preddbscan = dbscan.fit_predict(X)

# plot the clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_preddbscan)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Clustering with KMeans

KMeans take 2 parameters n_clusters(the amount of clusters to make) and random_state(to insure the result is reproducible)

```python
from sklearn.cluster import KMeans

# cluster the dataset using KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
y_predkmeans = kmeans.fit_predict(X)

# plot the clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_predkmeans)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Clustering with hierarchical clustering

Hierarchical clustering takes 2 parameters n_clusters(the amount of clusters to make) and linkage(the specific link criteria to merge the clusters)

there are 4 different linkage methods:

1. Single linkage: This method uses the distance between the closest points in two clusters as the distance between the clusters.

2. Complete linkage: This method uses the distance between the furthest points in two clusters as the distance between the clusters.

3. Average linkage: This method uses the average distance between all points in two clusters as the distance between the clusters.

4. Ward linkage: This method minimizes the variance of the distances between all points within a cluster.

```python
from sklearn.cluster import AgglomerativeClustering

# cluster the dataset using AgglomerativeClustering
agglom = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_predagglom = agglom.fit_predict(X)

# plot the clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_predagglom)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

to test the precision of each clustering, we can use the silhouette score for each

```python
from sklearn.metrics import silhouette_score

silhouettedbscan = silhouette_score(X, y_preddbscan)
silhouettekmeans = silhouette_score(X, y_predkmeans)
silhouetteagglom = silhouette_score(X, y_predagglom)

print(f'Silhouette score for DBSCAN: {silhouettedbscan}')
print(f'Silhouette score for KMeans: {silhouettekmeans}')
print(f'Silhouette score for AgglomerativeClustering: {silhouetteagglom}')
```
